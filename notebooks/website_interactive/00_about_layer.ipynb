# Notebook suite orientation (website layer)

**Layer A — non-canonical.** Excluded from `config/notebook_plan.json` and `config/expected_outputs.json`.

## Two-layer model

| Layer | Role | Typical files |
| ----- | ---- | ------------- |
| **A — Reader / interactive** | Markdown-only **story** notebooks (GitHub-safe) + **interactive** notebooks (ipywidgets; use **Voila** to hide inputs) | `*_story` paths below, `*_interactive` |
| **B — Technical** | Code cells + contract regeneration | `archival_shared/technical/*_execute.ipynb` |

## Where to start

- **Canonical narrative (no code):** `notebooks/archival_shared/01_smoke_test.ipynb`, `02_*.ipynb`, `03_*.ipynb`
- **Bounded interaction:** `*_interactive.ipynb` in the same folder (archival) or **`01_gate_explorer_interactive.ipynb`** here
- **Reproducibility:** `python reproduce_all.py` (executes **technical** notebooks only)

## Scientific integrity

Website demos call the **same public engine** as archival paths but **never** replace manifest validation.
