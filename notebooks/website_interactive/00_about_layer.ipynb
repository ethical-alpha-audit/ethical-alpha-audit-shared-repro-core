# NON-CANONICAL WEBSITE INTERACTIVE NOTEBOOK

## Website-facing layer (not part of archival validation)

This folder is **deliberately excluded** from `config/notebook_plan.json` and `config/expected_outputs.json`. Nothing here may replace the archival notebooks under `notebooks/archival_shared/`.

### Dependencies

- **ipywidgets** is *not* listed in `requirements.txt`. Sliders and dropdown widgets are therefore **not** used in the default environment. To add them, install `ipywidgets` in your environment separately; the maintainers must decide whether to adopt that dependency repo-wide.

### What you get instead

The companion notebooks use **matplotlib** and **pandas** (already in `requirements.txt`) plus **pre-defined case dictionaries** identical to the engine contracts. Readers switch scenarios by editing an integer index in a code cell and re-running — a simple, reviewable pattern that does not alter authoritative engine code.
